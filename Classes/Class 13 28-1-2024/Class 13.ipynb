```python
def add(num1, num2, *nums):
    print(num1)
    print(num2)
    print(*nums)
    added = num1 + num2
    
    for i in nums:    
        added += i
        
    print(added)
```

```python
add(3,4,5,6)
```

### Working with Flexible Number of Arguments with **

```python
def add(num1, num2, **nums):
    print(num1)
    print(num2)
    print(nums)
```

```python
add(3,4,5,6)
```

```python
add(3,4, num3=5, num4=6)
```

```python
def add(num1, num2, **nums):
    print(num1)
    print(num2)
    print(nums)
    added = num1 + num2
    for key, value in nums.items():
        added += value
    print(added)
```

```python
add(3,4, num3=5, num4=6)
```
